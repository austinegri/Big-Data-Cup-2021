```python
sequences = []

# for i, row in lag.iterrows():
    
#     if (i+1)%1000 == 0:
#         print(i)
#     if i == 10000:
#         break
#     last = lead.iloc[i]
    
#     # Played zone entries have pass, then a shot
#     if last['Event'] == 'Dump In/Out' and last['Detail 1'] == 'Retained': #and #(row['Event'] == 'Zone Entry' and row['Detail 1'] == 'Played'):
#         tmp.append(lead.iloc[i-2:i+3])
# pd.set_option('display.max_rows', 100)        
# tmp = pd.concat(tmp)
# tmp.head(30)

DEBUG = False

i = 1
last = df.iloc[0]
player = last['Player']
seq = []
l = len(df.index)
shot_assists = np.zeros(l)
goal_assists = np.zeros(l)
player2_pass_receiver = {'Play', 'Incomplete Play'}
last_pass = None
while i < l: #len(df.index):
    
    if (i+1) %10000 == 0:
        print("{}/{}".format(i-1, l))
    row = df.iloc[i]
    
    if last['GameId'] != row['GameId'] or last['Period'] != row['Period'] or (last['Home Team Skaters'] != row['Home Team Skaters']) or (last['Away Team Skaters'] != row['Away Team Skaters']) or (last['Home Team Goals'] != row['Home Team Goals']) or (last['Away Team Goals'] != row['Away Team Goals']):
        player = row['Player']
        if last_pass:
            if DEBUG : print("{}: Starting new Period/Game sequence... ({} - {}) {}".format(i, row['Event'], row['Detail 1'], row['Player']))
        last_pass = None
    
    ## TODO incorporate "Dump Out" Sequences ?
    elif row['Event'] in player2_pass_receiver: 
        last_pass = i
        player = row['Player 2']
        if DEBUG : print("{}: Starting pass sequence... ({} - {}) {} to {}".format(i, row['Event'], row['Detail 1'], row['Player'], row['Player 2']))
    elif row['Player'] == player:
        if last_pass:
            if DEBUG : print("\t{}: Continuing sequence w/ {} ({} - {})".format(i, row['Player'], row['Event'], row['Detail 1']))
        
        # same player event...
        if row['Event'] == 'Shot':
            if DEBUG : print("\t\tShot! Last Pass: {}".format(last_pass))
            if last_pass:
                shot_assists[last_pass] = 1
        elif row['Event'] == 'Goal':
            if DEBUG : print("\t\tGoal!! Last Pass: {}".format(last_pass))
            if last_pass:
                shot_assists[last_pass] = 1
                goal_assists[last_pass] = 1
    else:
        player = row['Player']
        if last_pass:
            if DEBUG : print("{}: Starting new sequence... ({} - {}) {}".format(i, row['Event'], row['Detail 1'], row['Player']))
        last_pass = None
    
    last = df.iloc[i]
    i += 1
    
print("done")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


passes = df[(df['Event'] == 'Play') | (df['Event'] == 'Incomplete Play')]
passes.filter(['Shot Assists', 'Goal Assists']).describe()
passes.filter(['Shot Assists', 'Goal Assists']).value_counts()
```

```python
X = passes.filter(['X Coordinate', 'Y Coordinate', 'X Coordinate 2', 'Y Coordinate 2', 'Detail 1'])

p1 = X.filter(['X Coordinate', 'Y Coordinate']).to_numpy()
p2 = X.filter(['X Coordinate 2', 'Y Coordinate 2']).to_numpy()
p1
X['Pass Dist'] = np.linalg.norm(p2 - p1, axis = 1)

X['ngle'] = (np.arctan2(p2[:,0] - p1[:,0], p2[:,1] - p1[:, 1])) / np.pi * 180


X.head(5)
```

```python


codes, uniques = pd.factorize(X['Detail 1'])
print(codes, uniques)
X['Detail 1'] = codes
y = passes.filter(['Goal Assists'])
X

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_test, y_test)
# reg.predict(X).argmax()
# df.iloc[7351]
```
